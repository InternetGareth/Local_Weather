```python
import pandas as pd
# Get well data




# Get well meta data
well_meta_data = pd.read_csv('well_meta_data.csv', index_col =0)
well_meta_data['lat'] = well_meta_data.lat.astype(float)
well_meta_data['lon'] = well_meta_data.lon.astype(float) * -1
well_meta_data['Site_ID'] = well_meta_data['well_tag_nr'].str.upper()
kitsap_meta_data = well_meta_data[well_meta_data['county_nm']=='Kitsap']
clean_kitsap_meta_data = kitsap_meta_data.dropna(subset = ['lat','lon','well_depth_qt','well_diameter_qt','well_log_id'])

# get well depths
well_depths=pd.read_csv("KPUD_Well_Depth_2020_04_05.csv", index_col=0)
well_depths.index = pd.to_datetime(well_depths.Date)
well_depths  = well_depths.drop('Date', axis=1)
well_depths['Site_ID'] = well_depths['Well Name'].apply(lambda x: x[:6]).str.upper()
well_depths['Date'] = well_depths.index.values


```

```python
import numpy as np
import re

all_text = open('scraped_positions.js')
line_list = all_text.readlines()

position_search = np.argwhere(['position' in x for x in line_list])
position_search = position_search.reshape(1,-1)[0]
lat_lons = [re.findall(r'[0-9][0-9.]+', line_list[x]) for x in position_search]
lats = [x[0] for x in lat_lons]
lons = [x[1] for x in lat_lons]

title_search = np.argwhere(['title' in x for x in line_list])
title_search = title_search.reshape(1,-1)[0]
names = [line_list[x].split("'")[1] for x in title_search]

names = pd.Series(names)
lats = pd.Series(lats)
lons = pd.Series(lons)
df_locations = pd.DataFrame([names, lats, lons]).transpose()
df_locations.columns = ['Name','Lat','Lon']
df_locations['Lon'] = df_locations['Lon'].astype(float)
df_locations['Lat'] = df_locations['Lat'].astype(float)
df_locations['Lon'] = df_locations['Lon'] * -1
```

```python
df_locations['Location_Key']=np.nan
df_locations['Location_Type']=np.nan

df_locations['Location_Key'][:40] = df_locations['Name'][:40].apply(lambda x: x[:2])
df_locations['Location_Type'][:40] = 'Discharge Station'

df_locations['Location_Key'][40:182] =  df_locations['Name'][40:182].apply(lambda x: x[:6])
df_locations['Location_Type'][40:182] ='Well'

df_locations['Location_Key'][182:236] = df_locations['Name'][182:236].apply(lambda x: x[:2])
df_locations['Location_Type'][182:236] ='Rain Gauge'
```

```python
df_locations.to_csv('KPUD_Locations.csv')
```
